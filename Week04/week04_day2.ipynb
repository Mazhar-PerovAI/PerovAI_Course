```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("perovskite_device_dataset.csv")
df
```

```python
X = df[
    ["Voc_V", "Jsc_mAcm-2", "FF", "Rs_ohm_cm2", "Rsh_ohm_cm2"]
]

y = df["PCE_percent"]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)

print("Actual PCE (%):   ", y.values)
print("Predicted PCE (%):", y_pred)
```

```python
coefficients = pd.Series(
    model.coef_,
    index=X.columns
)

coefficients
```

```python
coefficients.plot(kind="bar")
plt.ylabel("Model coefficient")
plt.title("Feature influence on PCE (Single‑device)")
plt.grid(False)
plt.show()
```

```python
coefficients.to_csv("pce_model_coefficients_day2.csv")
```
