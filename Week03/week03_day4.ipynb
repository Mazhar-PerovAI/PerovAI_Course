```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = {
    "Hours Studied": [1,2,3,4,5,6,7,8,9],
    "Sleep Hours":   [5,6,6,7,7,8,8,9,9],
    "Practice Tests": [0,1,1,2,2,3,3,4,4],
    "Previous Grade": [50,55,58,60,62,65,67,70,72],
    "Exam Score":    [52,55,60,63,66,70,74,78,82]
}

df = pd.DataFrame(data)
df
```

```python
corr = df.corr()
corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
